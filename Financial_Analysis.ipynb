```python
# Creating a dictionary to map full company names to their corresponding abbreviations from the A-share list.

fullname_to_name_dict = {item['fullname']: item['name'] for item in a_share_list}
print(fullname_to_name_dict)
```

```python
compete=pd.read_csv('/content/drive/My Drive/KnowledgeGraph/hidy.relationships.compete.csv')
dispute=pd.read_csv('/content/drive/My Drive/KnowledgeGraph/hidy.relationships.dispute.csv')
cooperate=pd.read_csv('/content/drive/My Drive/KnowledgeGraph/hidy.relationships.cooperate.csv')
invest=pd.read_csv('/content/drive/My Drive/KnowledgeGraph/hidy.relationships.invest.csv')
same_industry=pd.read_csv('/content/drive/My Drive/KnowledgeGraph/hidy.relationships.same_industry.csv')
supply=pd.read_csv('/content/drive/My Drive/KnowledgeGraph/hidy.relationships.supply.csv')
```

```python
supply.head()
```

```python
# Build opposite relationship dictionary
opposite = {}
for index, row in compete.iterrows():
    opposite.setdefault(row[':START_ID'], set()).add(row[':END_ID'])
    opposite.setdefault(row[':END_ID'], set()).add(row[':START_ID'])
for index, row in dispute.iterrows():
    opposite.setdefault(row[':START_ID'], set()).add(row[':END_ID'])
    opposite.setdefault(row[':END_ID'], set()).add(row[':START_ID'])

print(opposite)
```

```python
# Build same relationship dictionary.
same = {}
for index, row in cooperate.iterrows():
    same.setdefault(row[':START_ID'], set()).add(row[':END_ID'])
    same.setdefault(row[':END_ID'], set()).add(row[':START_ID'])
for index, row in same_industry.iterrows():
    same.setdefault(row[':START_ID'], set()).add(row[':END_ID'])
    same.setdefault(row[':END_ID'], set()).add(row[':START_ID'])
# While I believe 'supply' is directional, it's worth noting that companies in the upstream and downstream may have mutual impacts on each other.
for index, row in supply.iterrows():
    same.setdefault(row[':START_ID'], set()).add(row[':END_ID'])
    same.setdefault(row[':END_ID'], set()).add(row[':START_ID'])
## In investment relationships, the influence is expected to flow from the invested company to the investing company（END_ID to START_ID).
for index, row in invest.iterrows():
    same.setdefault(row[':END_ID'], set()).add(row[':START_ID'])

print(same)
```

```python
df.rename(columns={'Explicit_Company': 'ID'}, inplace=True)
merged_df = pd.merge(fv, df[['NewsID', 'ID']], on='NewsID', how='left')
```

```python
merged_df.head()
```

```python
def fill_columns(row, same, opposite):
    ids = row['ID']
    if row['label'] == 0:
        return [opposite.get(i, "N/A") for i in ids], [same.get(i, "N/A") for i in ids]
    else:
        return [same.get(i, "N/A") for i in ids], [opposite.get(i, "N/A") for i in ids]

merged_df[['Implicit_Positive_Company', 'Implicit_Negative_Company']] = merged_df.apply(lambda row: pd.Series(fill_columns(row, same, opposite)), axis=1)
```

```python
merged_df
```

```python
merged_df['Implicit_Positive_Company'] = merged_df['Implicit_Positive_Company'].astype(str).apply(lambda x: x.replace('{', '').replace('}', ''))
merged_df['Implicit_Negative_Company'] = merged_df['Implicit_Negative_Company'].astype(str).apply(lambda x: x.replace('{', '').replace('}', ''))
```

```python
merged_df['Implicit_Positive_Company'] = merged_df['Implicit_Positive_Company'].apply(ast.literal_eval)
merged_df['Implicit_Negative_Company'] = merged_df['Implicit_Negative_Company'].apply(ast.literal_eval)
```

```python
merged_df['Implicit_Positive_Company'] = merged_df['Implicit_Positive_Company'].apply(lambda x: list(set(x)))
merged_df['Implicit_Negative_Company'] = merged_df['Implicit_Negative_Company'].apply(lambda x: list(set(x)))
```
